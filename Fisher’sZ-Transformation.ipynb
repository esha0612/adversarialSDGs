```python
import numpy as np
import pandas as pd
import os

# 1. The 5 game r-values in order
r_values = [-0.28576158090903125, -0.37842697775326994, -0.38196900435707276, 0.029408546341634034, 0.08739401308940853]

# 2. List the paths to your 5 CSV files (one per game)
csv_paths = [
    '/content/correlation_analysis_aggregated_userLiarBars.csv',
    '/content/correlation_analysis_aggregated_user_branch2BOTC.csv',
    '/content/correlation_analysis_aggregated_user_branch3Paranoia.csv',
    '/content/correlation_analysis_aggregated_user_branch3SecretHitler.csv',
    '/content/correlation_analysis_aggregated_user_branchMafia.csv'
]

z_scores = []
total_weights = []

for r, path in zip(r_values, csv_paths):
    # Load the specific CSV for this game
    df = pd.read_csv(path)

    # Calculate the total sample size (N) for this game
    # This sums the Response_Count column you mentioned
    n_total = df['Response_Count'].sum()

    # Fisher z-transformation: z = arctanh(r)
    z = np.arctanh(r)

    # Weight for this game: (Total N - 3)
    weight = n_total - 3

    z_scores.append(z)
    total_weights.append(weight)

# 3. Calculate weighted average of z across all games
final_z = np.average(z_scores, weights=total_weights)

# 4. Back-transform to get the final Pooled Pearson r
final_pooled_r = np.tanh(final_z)

print(f"Overall Combined Correlation (r) for all 5 games: {final_pooled_r}")
```
